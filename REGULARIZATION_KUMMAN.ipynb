####  What is Regression?

Regression is a statistical technique used to model and analyze the relationship between a dependent variable (often denoted as y) and one or more independent variables (denoted as X). The most basic form is linear regression, where the relationship is modeled as a linear equation:

    y=a+bX
where a is the intercept and b is the slope (coefficient).

##### 1.Find the Optimal Value of a and b
Write a program in python to fit a linear curve for the
Given  dataset:

X=[1,2,3,4]

y=[2,4,6,9]

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 9])

# Perform linear regression
# np.polyfit returns the coefficients [a, b] for y = ax + b
coefficients = np.polyfit(x, y, 1)
a, b = coefficients

# Display the result
print(f"The fitted line is: y = {a:.2f}x + {b:.2f}")
print(f"a = {a:.2f}")
print(f"b = {b:.2f}")

# Plot the data points and the fitted line
plt.scatter(x, y, color='red', label='Data points')
plt.plot(x, a*x + b, label=f'Fitted line: y = {a:.2f}x + {b:.2f}', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Curve Fitting')
plt.show()
```

```python
import numpy as np

# Given data
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 9])

# Number of data points
n = len(x)

# Step 1: Calculate the sums
S_x = np.sum(x)         # Sum of x
S_y = np.sum(y)         # Sum of y
S_xy = np.sum(x * y)    # Sum of x * y
S_x2 = np.sum(x ** 2)   # Sum of x^2

# Step 2: Apply the formulas for a and b
a = (n * S_xy - S_x * S_y) / (n * S_x2 - S_x ** 2)
b = (S_y - a * S_x) / n

# Display the results
print(f"Sum of x (S_x) = {S_x}")
print(f"Sum of y (S_y) = {S_y}")
print(f"Sum of x*y (S_xy) = {S_xy}")
print(f"Sum of x^2 (S_x2) = {S_x2}\n")

print(f"Calculated slope (a) = {a:.2f}")
print(f"Calculated intercept (b) = {b:.2f}")

# Now you can predict the values of y for any x using y = ax + b
```

###### 2. What is Regularization?

Regularization is a technique used to prevent overfitting in a regression model by adding a penalty term to the loss function. This penalty discourages the model from becoming overly complex (i.e., using large coefficients). Common types of regularization include:

Lasso (L1 regularization): Adds the sum of the absolute values of the coefficients to the loss function.

Ridge (L2 regularization): Adds the sum of the squares of the coefficients to the loss function.

###### 3. What is a Regularization Parameter?

The regularization parameter (λ or c) controls the strength of the penalty. A larger value of λ makes the model simpler by reducing the size of the coefficients, while a smaller value allows the model to be more complex.

###### 4. Hyperparameters in Regression

Hyperparameters are parameters that are not learned from the data but are set before the training process. For regression, common hyperparameters include:

Regularization parameter (λ or c )
Learning rate (for iterative algorithms like gradient descent)
Maximum number of iterations

### Ridge Regression (L2 Regularization)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

# Given data
x = np.array([1, 2, 3, 4]).reshape(-1, 1)  # Reshaping to 2D array for sklearn
y = np.array([2, 4, 6, 9])


c = 10 

# Ridge Regression (L2 Regularization)
ridge_reg = Ridge(alpha=c)  
ridge_reg.fit(x, y)  

# Extract the coefficients
ridge_slope = ridge_reg.coef_[0]  # The slope (coefficient) of the fitted line
ridge_intercept = ridge_reg.intercept_  # The intercept of the fitted line


print(f"Ridge Regression (c={c}): y = {ridge_slope:.2f}x + {ridge_intercept:.2f}")

# Plot the data points and fitted lines
plt.scatter(x, y, color='red', label='Data points')  # Plot the original data points

# Plot Ridge line
plt.plot(x, ridge_slope * x + ridge_intercept, label=f'Ridge: y = {ridge_slope:.2f}x + {ridge_intercept:.2f}', color='blue')  # Plot the fitted line

plt.xlabel('x')  
plt.ylabel('y')  
plt.legend()  
plt.title(f'Ridge Regression (c={c})')  
plt.show()  
```

```python
import numpy as np
from sklearn.linear_model import Ridge

# Define the dataset
X = np.array([1, 2, 3, 4]).reshape(-1, 1)  # Reshape X to be a 2D array for sklearn
y = np.array([2, 4, 6, 9])


lambda_value = 10

# Create the Ridge regression model with the regularization parameter
ridge = Ridge(alpha=lambda_value)

# Fit the model to the dataset
ridge.fit(X, y)

intercept = ridge.intercept_  # Intercept (w0)
coefficient = ridge.coef_[0]  # Coefficient (w1)

# Display the mathematical formula with explanations
print("Formula Used in Ridge Regression: J(w0, w1) = Σ (yi - (w0 + w1 * xi))^2 + λ * w1^2")
print(f"Where λ (regularization parameter) = {lambda_value}")
print(f"Minimizing the loss function with regularization term (λ * w1^2):")

# Output the calculated intercept and coefficient
print(f"Intercept (w0): {intercept}")
print(f"Coefficient (w1): {coefficient}")

# Display the resulting equation of the line
print(f"\nThe equation of the fitted line is: y = {intercept:.2f} + {coefficient:.2f} * X")

# Now, let's use the model to predict some values (just for visualization)
y_pred = ridge.predict(X)
print("\nPredicted values for the given X:")
for xi, yi_pred in zip(X.flatten(), y_pred):
    print(f"X = {xi}: Predicted y = {yi_pred:.2f}")
```

```python
import numpy as np

# Dataset
X = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 9])

# Regularization parameter (lambda)
lambda_value = 10


x_mean = np.mean(X)
y_mean = np.mean(y)


numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean)**2) + lambda_value

# Compute the coefficient w1 (slope)
w1 = numerator / denominator

# Compute the intercept w0
w0 = y_mean - w1 * x_mean

# Output the results
print(f"Coefficient (w1): {w1}")
print(f"Intercept (w0): {w0}")

# Display the resulting equation
print(f"The equation of the fitted line is: y = {w0:.2f} + {w1:.2f} * X")

plt.scatter(X, y, color='red', label='Data Points')

# Predicted y values using the line equation
y_pred = w0 + w1 * X

# Plotting the regression line
plt.plot(X, y_pred, color='blue', label=f'Fitted Line: y = {w0:.2f} + {w1:.2f}*x')

# Adding labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Ridge Regression Line')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

### Lasso Regression (L1 Regularization)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso

# Given data
x = np.array([1, 2, 3, 4]).reshape(-1, 1)
y = np.array([2, 4, 6, 9])


# Lasso Regression (L1 Regularization)
lasso_reg = Lasso(alpha=10)  # alpha is the regularization parameter (λ)
lasso_reg.fit(x, y)
lasso_slope = lasso_reg.coef_[0]
lasso_intercept = lasso_reg.intercept_

# Display the results
print(f"Lasso Regression: y = {lasso_slope:.2f}x + {lasso_intercept:.2f}")

# Plot the data points and fitted lines
plt.scatter(x, y, color='red', label='Data points')


# Plot Lasso line
plt.plot(x, lasso_slope * x + lasso_intercept, label=f'Lasso: y = {lasso_slope:.2f}x + {lasso_intercept:.2f}', color='green')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title(' Lasso Regularization')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
X = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 9])

# Regularization parameter (lambda)
lambda_value = 10

# Compute the means of X and y
x_mean = np.mean(X)
y_mean = np.mean(y)

# Compute the numerator and denominator for w1 (slope)
numerator = np.sum((X - x_mean) * (y - y_mean)) - lambda_value
denominator = np.sum((X - x_mean)**2)


w1 = numerator / denominator


w1 = np.sign(w1) * max(0, abs(numerator)-lambda_value) / denominator


w0 = y_mean - w1 * x_mean

# Output the results
print(f"Coefficient (w1): {w1}")
print(f"Intercept (w0): {w0}")

# Display the resulting equation
print(f"The equation of the fitted line is: y = {w0:.2f} + {w1:.2f} * X")


plt.scatter(X, y, color='red', label='Data Points')

# Predicted y values using the line equation
y_pred = w0 + w1 * X

# Plotting the regression line
plt.plot(X, y_pred, color='blue', label=f'Fitted Line: y = {w0:.2f} + {w1:.2f}*x')

# Adding labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Lasso Regression Line')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score


X = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 9])

lambda_ridge = 10
lambda_lasso = 10

# Compute means of X and y
x_mean = np.mean(X)
y_mean = np.mean(y)


def calculate_ridge_coefficients(X, y, lambda_ridge):
    numerator = np.sum((X - x_mean) * (y - y_mean))
    denominator = np.sum((X - x_mean)**2)+ lambda_ridge
    w1 = numerator / denominator
    w0 = y_mean - w1 * x_mean
    return w0, w1


def calculate_lasso_coefficients(X, y, lambda_lasso):
    numerator = np.sum((X - x_mean) * (y - y_mean)) - lambda_lasso
    denominator = np.sum((X - x_mean)**2)
    w1 = np.sign(numerator) * max(0, abs(numerator) - lambda_lasso) / denominator
    w0 = y_mean - w1 * x_mean
    return w0, w1


ridge_w0, ridge_w1 = calculate_ridge_coefficients(X, y, lambda_ridge)
lasso_w0, lasso_w1 = calculate_lasso_coefficients(X, y, lambda_lasso)


ridge_y_pred = ridge_w0 + ridge_w1 * X
lasso_y_pred = lasso_w0 + lasso_w1 * X


ridge_mse = mean_squared_error(y, ridge_y_pred)
ridge_r2 = r2_score(y, ridge_y_pred)


lasso_mse = mean_squared_error(y, lasso_y_pred)
lasso_r2 = r2_score(y, lasso_y_pred)


print("Ridge Regression:")
print(f"Coefficients (w0, w1): ({ridge_w0:.2f}, {ridge_w1:.2f})")
print(f"MSE: {ridge_mse:.2f}, R2 score: {ridge_r2:.2f}")

print("\nLasso Regression:")
print(f"Coefficients (w0, w1): ({lasso_w0:.2f}, {lasso_w1:.2f})")
print(f"MSE: {lasso_mse:.2f}, R2 score: {lasso_r2:.2f}")


plt.figure(figsize=(10, 6))


plt.scatter(X, y, color='red', label='Data Points')


plt.plot(X, ridge_y_pred, color='blue', linestyle='-', label=f'Ridge Regression: y = {ridge_w0:.2f} + {ridge_w1:.2f} * X')


plt.plot(X, lasso_y_pred, color='green', linestyle='--', label=f'Lasso Regression: y = {lasso_w0:.2f} + {lasso_w1:.2f} * X')


plt.xlabel('X')
plt.ylabel('y')
plt.title('Comparison of Ridge and Lasso Regression')
plt.legend()


plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score

# Dataset
X = np.array([1, 2, 3, 4]).reshape(-1, 1)
y = np.array([2, 4, 6, 9])

# Regularization parameter
alpha = 10  # This is λ in Ridge and Lasso

# Ridge Regression using scikit-learn
ridge_model = Ridge(alpha=alpha)
ridge_model.fit(X, y)
ridge_y_pred = ridge_model.predict(X)

# Lasso Regression using scikit-learn
lasso_model = Lasso(alpha=alpha)
lasso_model.fit(X, y)
lasso_y_pred = lasso_model.predict(X)

# Calculate MSE and R2 score for Ridge regression
ridge_mse = mean_squared_error(y, ridge_y_pred)
ridge_r2 = r2_score(y, ridge_y_pred)

# Calculate MSE and R2 score for Lasso regression
lasso_mse = mean_squared_error(y, lasso_y_pred)
lasso_r2 = r2_score(y, lasso_y_pred)

# Print coefficients, MSE, and R2 score for Ridge
print("Ridge Regression:")
print(f"Coefficients (Intercept, Slope): ({ridge_model.intercept_:.2f}, {ridge_model.coef_[0]:.2f})")
print(f"MSE: {ridge_mse:.2f}, R2 score: {ridge_r2:.2f}")

# Print coefficients, MSE, and R2 score for Lasso
print("\nLasso Regression:")
print(f"Coefficients (Intercept, Slope): ({lasso_model.intercept_:.2f}, {lasso_model.coef_[0]:.2f})")
print(f"MSE: {lasso_mse:.2f}, R2 score: {lasso_r2:.2f}")

# Plotting
plt.figure(figsize=(10, 6))

# Plotting data points
plt.scatter(X, y, color='red', label='Data Points')

# Plotting Ridge regression line
plt.plot(X, ridge_y_pred, color='blue', label=f'Ridge Regression: y = {ridge_model.intercept_:.2f} + {ridge_model.coef_[0]:.2f} * X')

# Plotting Lasso regression line
plt.plot(X, lasso_y_pred, color='green', linestyle='--', label=f'Lasso Regression: y = {lasso_model.intercept_:.2f} + {lasso_model.coef_[0]:.2f} * X')

# Adding labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Comparison of Ridge and Lasso Regression')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
